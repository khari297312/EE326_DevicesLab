```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Id_35 = np.zeros(401, dtype=np.float64)
Vd_35 = np.zeros(401, dtype=np.float64)
Id_45 = np.zeros(401, dtype=np.float64)
Vd_45 = np.zeros(401, dtype=np.float64)
Id_55 = np.zeros(401, dtype=np.float64)
Vd_55 = np.zeros(401, dtype=np.float64)
Id_65 = np.zeros(401, dtype=np.float64)
Vd_65 = np.zeros(401, dtype=np.float64)
Id_75 = np.zeros(401, dtype=np.float64)
Vd_75 = np.zeros(401, dtype=np.float64)
```

```python
Il_35 = np.zeros(401, dtype=np.float64)
Vl_35 = np.zeros(401, dtype=np.float64)
Il_45 = np.zeros(401, dtype=np.float64)
Vl_45 = np.zeros(401, dtype=np.float64)
Il_55 = np.zeros(401, dtype=np.float64)
Vl_55 = np.zeros(401, dtype=np.float64)
Il_65 = np.zeros(401, dtype=np.float64)
Vl_65 = np.zeros(401, dtype=np.float64)
Il_75 = np.zeros(401, dtype=np.float64)
Vl_75 = np.zeros(401, dtype=np.float64)
```

```python
with open("dark_35.dat") as d35:
    for idx, line in enumerate(d35):
        a, i, b, v = line.split()
        Id_35[idx] = float(i)
        Vd_35[idx] = float(v)

with open("dark_45.dat") as d45:
    for idx, line in enumerate(d45):
        a, i, b, v = line.split()
        Id_45[idx] = float(i)
        Vd_45[idx] = float(v)

with open("dark_55.dat") as d55:
    for idx, line in enumerate(d55):
        a, i, b, v = line.split()
        Id_55[idx] = float(i)
        Vd_55[idx] = float(v)

with open("dark_65.dat") as d65:
    for idx, line in enumerate(d65):
        a, i, b, v = line.split()
        Id_65[idx] = float(i)
        Vd_65[idx] = float(v)

with open("dark_75.dat") as d75:
    for idx, line in enumerate(d75):
        a, i, b, v = line.split()
        Id_75[idx] = float(i)
        Vd_75[idx] = float(v)
```

```python
with open("light_35.dat") as d35:
    for idx, line in enumerate(d35):
        a, i, b, v = line.split()
        Il_35[idx] = float(i)
        Vl_35[idx] = float(v)

with open("light_45.dat") as d45:
    for idx, line in enumerate(d45):
        a, i, b, v = line.split()
        Il_45[idx] = float(i)
        Vl_45[idx] = float(v)

with open("light_55.dat") as d55:
    for idx, line in enumerate(d55):
        a, i, b, v = line.split()
        Il_55[idx] = float(i)
        Vl_55[idx] = float(v)

with open("light_65.dat") as d65:
    for idx, line in enumerate(d65):
        a, i, b, v = line.split()
        Il_65[idx] = float(i)
        Vl_65[idx] = float(v)

with open("light_75.dat") as d75:
    for idx, line in enumerate(d75):
        a, i, b, v = line.split()
        Il_75[idx] = float(i)
        Vl_75[idx] = float(v)
```

```python
plt.plot(Vd_35, Id_35, label="$35^{\circ} C$")
plt.plot(Vd_45, Id_45, label="$45^{\circ} C$")
plt.plot(Vd_55, Id_55, label="$55^{\circ} C$")
plt.plot(Vd_65, Id_65, label="$65^{\circ} C$")
plt.plot(Vd_75, Id_75, label="$75^{\circ} C$")
plt.xlabel("V")
plt.ylabel("mA")
plt.title("Plot of Dark IV at different Temperatures")
plt.grid(True)
plt.legend()
```

```python
plt.plot(Vl_35[:243], -Il_35[:243], label="$35^{\circ} C$")
plt.plot(Vl_45[:241], -Il_45[:241], label="$45^{\circ} C$")
plt.plot(Vl_55[:238], -Il_55[:238], label="$55^{\circ} C$")
plt.plot(Vl_65[:236], -Il_65[:236], label="$65^{\circ} C$")
plt.plot(Vl_75[:233], -Il_75[:233], label="$75^{\circ} C$")
plt.xlabel("V")
plt.ylabel("mA")
plt.title("Plot of Light IV at different Temperatures")
plt.grid(True)
plt.legend()
```
